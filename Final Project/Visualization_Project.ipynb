```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.io as pio
from dash import Dash,html,dcc
from jupyter_dash import JupyterDash 
from dash.dependencies import Input,Output,State
import io
from google.colab import files
from sklearn import tree
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import dash_bootstrap_components as dbc
```

```python
uploaded = files.upload()
```

```python
data = pd.read_csv(io.BytesIO(uploaded[r"cleaned_data3.csv"]))
data.head()
```

### Data Preparation

```python
data.rename(columns={'male':'Gender'},inplace=True)
data.rename(columns={'TenYearCHD':'jeopardy of having heart disease'},inplace=True)
data['Gender'].replace(0, 'Female',inplace=True)
data['Gender'].replace(1, 'Male',inplace=True)
data['currentSmoker'].replace(0, 'No',inplace=True)
data['currentSmoker'].replace(1, 'Yes',inplace=True)
data=data.sort_values(by="cigsPerDay",ascending=True)
data.head(10)
```

```python
data.info()
```

### Getting Data Ready For Dashboard

```python
data['currentSmoker'].unique()
```

```python
data_male = data[data['Gender'] == 'Male']
data_female = data[data['Gender'] == 'Female']
```

```python
smoke_df = data.groupby(['currentSmoker','Gender'])[['jeopardy of having heart disease']].count().reset_index()
smoke_df
```

```python
diabetes_df = data.groupby(['age','Gender'])[['diabetes']].count().reset_index()
diabetes_df.head()
```

```python
hyp_df = data.groupby(['age'])[['prevalentHyp']].count().reset_index()
hyp_df.head()
```

```python
bp_df = data.groupby(['age'])[['BPMeds']].count().reset_index()
bp_df.head()
```

```python
stroke_df = data.groupby(['age'])[['prevalentStroke']].count().reset_index()
stroke_df.head()
```

```python
sysBP_df = data.groupby(['Gender','currentSmoker'])[['sysBP']].count().reset_index().rename(columns={"sysBP":"Systolic Blood Pressure"})
sysBP_df
```

```python
diaBP_df = data.groupby(['Gender','currentSmoker'])[['diaBP']].count().reset_index().rename(columns={"diaBP":"Diastolic Blood Pressure"})
diaBP_df
```

```python
chol_df = data.groupby(['Gender','currentSmoker'])[['totChol']].count().reset_index().rename(columns={"totChol":"Cholesterol"})
chol_df
```

### Dashboard Code

```python
female = data['Gender'].value_counts()[0]
male = data['Gender'].value_counts()[1]
smokers = data['currentSmoker'].value_counts()[1]
totalHD = data['jeopardy of having heart disease'].value_counts()[1]
```

```python
from plotly.graph_objs import *

layout = go.Layout(
            paper_bgcolor='rgba(0,0,0,0)',
            plot_bgcolor='rgba(0,0,0,0)'
        )
```

```python
fig1 = px.scatter(data,x='heartRate',y='BMI',color='Gender',animation_frame='cigsPerDay',size_max=70,hover_name='Gender',title='The smoking effect on our heart rate and BMI',width=950,height =640, template='plotly_dark')

fig14 = px.pie(data,names=data['Gender'],values=data['prevalentStroke'],title='Stroke vs Gender',width=410,height =400, template='plotly_dark')
fig15 = px.pie(data,names=data['Gender'],values=data['prevalentHyp'],title='Hyper Tension vs Gender',width=410,height =400, template='plotly_dark')

fig4 = px.bar(chol_df, x="Gender", y="Cholesterol", color="currentSmoker", barmode="group",title='The smoking effect on our Cholesterol.',width=410,height =400, template='plotly_dark')
fig5 = px.bar(diaBP_df, x="Gender", y="Diastolic Blood Pressure", color="currentSmoker", barmode="group",title='The smoking effect on our Blood Pressure.',width=400,height =400, template='plotly_dark')


fig7 = make_subplots(rows=3, cols=1,
                    vertical_spacing=0.1,
                    horizontal_spacing=0.5,x_title="Age", subplot_titles=['Its effect on our brain','Its effect on our heart efficiency' , 'Its effect on our blood pressure' ,'Its effect on our blood sugar level'])

#fig7.add_trace(go.Bar(x=data["age"], y=data["diabetes"],name="Diabetes"),
 #             row=4, col=1)

fig7.add_trace(go.Bar(x=data["age"], y=data["BPMeds"],name="Blood Pressure"),
              row=3, col=1)

fig7.add_trace(go.Bar(x=data["age"], y=data["prevalentHyp"],name="Hyper Tension"),
              row=2, col=1)

fig7.add_trace(go.Bar(x=data["age"], y=data["prevalentStroke"],name="Stroke"),
              row=1, col=1)

fig7.update_layout(height=600, width=900,title='The aging effect on our health.', template='plotly_dark')

```

```python
app = JupyterDash(external_stylesheets=[dbc.themes.CYBORG])
app.layout = html.Div([
                       
          html.Div([
                   html.H1("Health Dashboard"),    
          ],style = {'textAlign': 'center'}),

          html.Div([
              dbc.Row([
                       dbc.Card([
                          dbc.CardBody([
                                        html.H4("Males", className="card-title"),
                                        html.H4(str(male), className="card-title")],style = {'textAlign': 'center', 'font-weight': 'bold'})]),
                      dbc.Card([
                          dbc.CardBody([
                                      html.H4("Females", className="card-title"),
                                      html.H4(str(female), className="card-title")],style = {'textAlign': 'center', 'font-weight': 'bold'},
                          ),

                       ]),
                     dbc.Card([
                          dbc.CardBody([
                                        html.H4("Smokers", className="card-title"),
                                        html.H4(str(smokers), className="card-title")],style = {'textAlign': 'center', 'font-weight': 'bold', 'font_size': '90px'}),

                       ]),
                     dbc.Card([
                          dbc.CardBody([                                      
                                        html.H5("Total Heart Diseases Patients", className="card-title"),
                                        html.H4(str(totalHD), className="card-title"),],style = {'textAlign': 'center', 'font-weight': 'bold', 'font_size': '90px'}),
                       ])
          ],style = { 'height': 200}),
          

          ],style = { 'columnCount': 4, 'padding': 8}),
         
          html.Div([
              dbc.Row([
                        dbc.Col([
                          dcc.Graph(id='my_graph1',figure=fig1),                                       
                        ]),
                       dbc.Col([
                                                       dcc.Dropdown(
                                        id='dropdown',
                                        options=[{'label': str(gender), 'value': str(gender)}  for gender in data['Gender'].unique()
                                        ],
                                        value=None,                                    
                                        placeholder='Choose a gender',
                                        
                                ),
                          dcc.Graph(id='my_graph7',figure=fig7),                                       
                        ])
          ]),
          ]),
          html.Div([
              dbc.Row([
                       dbc.Col([
                        dcc.Graph(id='my_graph14',figure=fig14),
                                               ]),
                      dbc.Col([
                        dcc.Graph(id='my_graph15',figure=fig15),

                       ]),
                     dbc.Col([
                        dcc.Graph(id='my_graph4',figure=fig4)

                       ]),
                     dbc.Col([
                        dcc.Graph(id='my_graph5',figure=fig5)

                       ])
          ],style = { 'columnCount': 4}),
          

          ],style = { 'columnCount': 2, 'padding': 8})
                               
])

@app.callback(
    Output('my_graph7','figure'),
    Input('dropdown','value'))   
def update_graph(dropdownvalue):
    if dropdownvalue==None:
      fig7 = make_subplots(rows=3, cols=1,
                    vertical_spacing=0.1,
                    horizontal_spacing=0.5,x_title="Age", subplot_titles=['Its effect on our brain','Its effect on our heart efficiency' , 'Its effect on our blood pressure' ,'Its effect on our blood sugar level'])
      
      fig7.add_trace(go.Bar(x=data["age"], y=data["BPMeds"],name="Blood Pressure"),
              row=3, col=1)
      fig7.add_trace(go.Bar(x=data["age"], y=data["prevalentHyp"],name="Hyper Tension"),
              row=2, col=1)
      fig7.add_trace(go.Bar(x=data["age"], y=data["prevalentStroke"],name="Stroke"),
              row=1, col=1)
      fig7.update_layout(height=600, width=900,title='The aging effect on our health.', template='plotly_dark')
    elif dropdownvalue == 'Male':
      fig7 = make_subplots(rows=3, cols=1,
                    vertical_spacing=0.1,
                    horizontal_spacing=0.5,x_title="Age", subplot_titles=['Its effect on our brain','Its effect on our heart efficiency' , 'Its effect on our blood pressure' ,'Its effect on our blood sugar level'])
      
      fig7.add_trace(go.Bar(x=data_male["age"], y=data_male["BPMeds"],name="Blood Pressure"),
              row=3, col=1)
      fig7.add_trace(go.Bar(x=data_male["age"], y=data_male["prevalentHyp"],name="Hyper Tension"),
              row=2, col=1)
      fig7.add_trace(go.Bar(x=data_male["age"], y=data_male["prevalentStroke"],name="Stroke"),
              row=1, col=1)
      fig7.update_layout(height=600, width=900,title='The aging effect on our health.', template='plotly_dark')
    return fig7

app.run_server() #Open the link below to check the output
```
